# Q2. Python If-Else

```python
n=int(input().strip())
if n%2!=0:
    print('Weird')
elif n%2==0 and n in range(2,6):
    print('Not Weird')
elif n%2==0 and n in range(6,21):
    print('Weird')
elif n%2==0 and n>20:
    print('Not Weird')
```

# Q6. Write a function

```python
def is_leap(year):
    leap=False
    if year%4==0 and (year%100!=0 or year%400==0):
        leap=True
    return leap

is_leap(1900)
```
