# Defanging an IP Address

```python
def defangIPaddr(address):
    result=address.replace('.','[.]')
    return result
print(defangIPaddr("1.1.1.1"))
```

# Find Numbers with Even Number of Digits

```python
def findnumber(num):
    count=0
    for i in num:
        digits=0
        temp=i
        while temp>0:
            digits+=1
            temp//=10
        if digits%2==0:
            count+=1
    return count
print(findnumber([12,345,2,6,7896]))
```

# Number of Good Pairs

```python
def numIdenticalPairs(nums):
    freq={}
    count=0
    for num in nums:
        count+=freq.get(num,0)
        freq[num]= freq.get(num,0)+1
    return count
print(numIdenticalPairs([1,2,3,1,1,3]))
```

# How Many Numbers Are Smaller Than the Current Number

```python
def smallerNumberThanCurrent(nums):
    freq=[0]*101
    for num in nums:
        freq[num]+=1
    for i in range(1,101):
        freq[i]+=freq[i-1]
    return [freq[num-1] if num >0 else 0 for num in nums]
print(smallerNumberThanCurrent([8,1,2,2,3]))
```

# Subtract the Product and Sum of Digits of an Integer

```python
def subProdandSum(n):
    product=1
    total_sum=0
    while n>0:
        digit=n%10
        product*=digit
        total_sum+=digit
        n//=10
    return product-total_sum
print(subProdandSum(234))
```

# XOR Operation in an Array

```python
def xorOp(n,start):
    res=0
    for i in range(n):
        res^=start+2*i
    return res
print(xorOp(n=5,start=0))
```
