# Running Sum of 1D Array

```python
def runningsum(nums):
    result=[]
    total_sum=0
    for i in nums:
        total_sum+=i
        result.append(total_sum)
    return result

print(runningsum(nums=[1,2,3,4,5]))
```

# Shuffle the Array

```python
def shuffle(nums,n):
    result=[]
    temp1=nums[:n]
    temp2=nums[n:]
    for a,b in zip(temp1,temp2):
        result.append(a)
        result.append(b)
    return result



print(shuffle([1,2,3,4,5,6],3))
```

# Kids With the Greatest Number of Candies

```python
def kidswithCandies(candies,extracandies):
    maximum=max(candies)
    result=[]
    for i in candies:
        if (i + extracandies) >= maximum:
            result.append(True)
        else:
            result.append(False)
    return result

print(kidswithCandies([2,3,5,1,3],3))
```
